```python
import pandas as pd
import numpy as np
```

```python
csv_file_path = 'pred_wrong_report.csv'
df = pd.read_csv(csv_file_path)
```

```python
df_arr = df.values
fname_arr = []
c = 0
for fn in df_arr[:,0]: 
    if fn.split('_')[0] == str(c):
        fname_arr.append(fn[1+len(str(c)):-6])
#         data = [c, fn[2:-6]]
#         fname_arr.append(data)
        c += 1
fname_arr
```

```python
thresh_list = np.ones(380) * 0.8

for d in df_arr:
    if str(d[1]) != 'nan':
        wrong_tags = str(d[1]).split('\n')
#         print(wrong_tags)
        for wrong_tag in wrong_tags:
#             print(wrong_tag.split(" ")[0])
            cls_idx, = np.where(wrong_tag.split(" ")[0] == np.array(fname_arr))
#             print(cls_idx)
            cls_idx = int(cls_idx[0])
#             print(cls_idx)
            prob = float(wrong_tag.split(" ")[1][1:-2])
#             print(prob)
            if thresh_list[cls_idx] < prob:
                thresh_list[cls_idx] = prob
thresh_list
```

```python
# df_output = pd.DataFrame(data=thresh_list, index=np.arange(380), columns=['ID','Threshold'])
mtx_output = []
for i, t in enumerate(thresh_list):
    mtx_output.append([i,t])
    
df_output = pd.DataFrame(data=mtx_output, columns=['ID','Threshold'])
df_output
```
