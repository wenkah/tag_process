```python
import pandas as pd
import numpy as np
```

```python
csv_file_path = '/Users/wenkai/Desktop/pred_wrong_report.csv'
df = pd.read_csv(csv_file_path)
```

```python
df_arr = df.values
cls_dict = {}
index = 0
for name in df_arr[:,0]: 
    name_split = name.split('_')[0]
    if int(name_split) == index:
        cls_dict[name[len(str(index))+1:-6]] = index
        index += 1
cls_dict
```

```python
thresh_list = np.ones(380) * 0.8

for item in df_arr:
    if str(item[1]) != 'nan':
        wrong_tags = str(item[1]).split('\n')
        for wrong_tag in wrong_tags:
            wrong_item = wrong_tag.split(" ")[0]
            wrong_item_prob = float(wrong_tag.split(" ")[1][1:-2])
            for cls in cls_dict:
                if wrong_item == cls:
                    cls_idx = cls_dict[cls]
            if thresh_list[cls_idx] < wrong_item_prob:
                thresh_list[cls_idx] = wrong_item_prob
thresh_list
```

```python
# df_output = pd.DataFrame(data=thresh_list, index=np.arange(380), columns=['ID','Threshold'])
mtx_output = []
for i, t in enumerate(thresh_list):
    mtx_output.append([i,t])
    
df_output = pd.DataFrame(data=mtx_output, columns=['ID','Threshold'])
df_output.to_csv('/Users/wenkai/Desktop/tag_process/tags_output', sep='\t', encoding='utf-8')
```
